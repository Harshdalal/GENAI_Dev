```python
# Import necessary libraries from Hugging Face
from transformers import MarianMTModel, MarianTokenizer

# Define the source and target languages (e.g., English to French)
source_lang = 'en'
target_lang = 'fr'

# Load pre-trained MarianMT model and tokenizer for the specified languages
model_name = f'Helsinki-NLP/opus-mt-{source_lang}-{target_lang}'
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

# Function to translate text
def translate(text, source_lang, target_lang):
    # Tokenize the input text
    tokens = tokenizer.prepare_seq2seq_batch(src_texts=[text], return_tensors="pt")

    # Perform the translation
    translated_tokens = model.generate(**tokens)

    # Decode the translated tokens into text
    translated_text = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

    return translated_text


```

```python
# Example Usage
source_text = "Hello, how are you haras dalal?"
translated_text = translate(source_text, source_lang, target_lang)

print(f"Original: {source_text}")
print(f"Translated: {translated_text}")
```
